```python
import pandas as pd

# Read the Excel file
file_path = 'quay.xlsx'
df_ship = pd.read_excel(file_path, sheet_name='ship', header=0, index_col=None)

# Collect ship names from the A column
ship_names = df_ship['Ship_Name'].values.tolist()
```

```python

# Initialize a dictionary to store operation lists for each ship
ship_operations = {}
ship_type = {}

# Iterate through each ship and collect corresponding operation names
for ship_name in ship_names:
    # Filter rows where 'A' column matches the current ship name
    ship_data = df_ship[df_ship['Ship_Name'] == ship_name]
    
    # Collect operation names from the I-th column
    operation_names = ship_data['Operation_Type'].values.tolist()
    type_name = ship_data['Ship_Type'].values.tolist()
    
    # Store operation names in the dictionary
    ship_operations[ship_name] = []
    for i in range(len(operation_names)):
        ship_operations[ship_name].append((type_name[i],operation_names[i]))

# Print or use the collected data as needed
# print("Ship Names:", ship_names)
print("Ship Operations:", ship_operations)
```

```python
df_quay = pd.read_excel(file_path, sheet_name='quay', header=0, index_col=None)
quay_list = df_quay.columns.tolist()
quay_list = quay_list[2:]
```

```python

quay_data = {}
for index, row in df_quay.iterrows():
    shiptype = row['선종']
    worktype = row['작업']
    
    if (shiptype, worktype) not in quay_data:
        quay_data[(shiptype, worktype)] = {}
    for q in quay_list:
        priority = row[q]
        if priority != 'N':
            if priority not in quay_data[(shiptype, worktype)]:
                quay_data[(shiptype, worktype)][priority]=[q]
            else:
                quay_data[(shiptype, worktype)][priority].append(q)
```

```python
job_list = df_ship['Ship_Name'].unique().tolist()
ship_quay = {}
for job in job_list:
    op_list = ship_operations[job]
    ship_quay[job] = []
    for op in op_list:
        ship_quay[job].append(quay_data[op])
    
ship_quay
```

```python
result_dict = {}
machine_list = ['M1', 'M2','M3']
for index, row in df.iterrows():
    shiptype = row['선종']
    worktype = row['작업']
    
    if (shiptype, worktype) not in result_dict:
        result_dict[(shiptype, worktype)] = {}
    for q in machine_list:
        priority = row[q]
        if priority != 'N':
            if str(int(priority)) not in result_dict[(shiptype, worktype)]:
                result_dict[(shiptype, worktype)][str(int(priority))]=[q]
            else:
                result_dict[(shiptype, worktype)][str(int(priority))].append(q)
print(result_dict)
```
